```python
import pandas as pd
```

```python
df = pd.read_json("openbeelden.json")
attrs = pd.DataFrame.from_records(df.attributes)
# list_attributes seems to be `locations` or `subject` parsed into lists 
#      (they are comma-separated strings)
# list_attrs = pd.DataFrame.from_records(df.list_ttributes)
df = pd.concat([df, attrs], axis=1)  # df = pd.concat([df, attrs, list_attrs], axis=1)
df = df.drop(["attributes", "list_attributes"], axis=1)

df["EndDate"] = df.date
df["DateString"] = df.date

column_mapping = {
    "id": "ID",
    "name": "Title",
    "description": "Description",
    "date": "StartDate"
}
df = df.rename(column_mapping, axis=1)

img_cols = ["image_url", "thumbnail_url"]
img_df = df[["ID"]+img_cols].copy()
df = df.drop(img_cols, axis=1)
img_df.to_csv("image_URLs.csv")
```
